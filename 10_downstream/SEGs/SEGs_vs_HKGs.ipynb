```python
import pandas as pd
```

```python
def compare_seg_hkg(seg_file, hkg_file, species):

    print(f"\n====== {species} ======")

    # 读取数据
    seg_df = pd.read_csv(seg_file)
    hkg_df = pd.read_csv(hkg_file, sep=";")

    # 提取基因集合
    seg_genes = set(seg_df["gene"])
    hkg_genes = set(hkg_df.iloc[:, 1])  # 第二列是gene
    print(len(hkg_genes))
    # 交集
    overlap = seg_genes & hkg_genes

    # 统计
    n_seg = len(seg_genes)
    n_hkg = len(hkg_genes)
    n_overlap = len(overlap)

    # 比例
    overlap_ratio_seg = n_overlap / n_seg
    overlap_ratio_hkg = n_overlap / n_hkg

    # Jaccard
    jaccard = n_overlap / len(seg_genes | hkg_genes)

    # 输出
    print(f"SEG数量: {n_seg}")
    print(f"HKG数量: {n_hkg}")
    print(f"重叠基因数: {n_overlap}")
    print(f"Overlap / SEG: {overlap_ratio_seg:.3f}")
    print(f"Overlap / HKG: {overlap_ratio_hkg:.3f}")
    print(f"Jaccard index: {jaccard:.3f}")

    return overlap
```
