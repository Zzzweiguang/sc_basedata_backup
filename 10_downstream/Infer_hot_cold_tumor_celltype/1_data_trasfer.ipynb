# GSE115978

```python
import pandas as pd
from pathlib import Path

data_dir = Path("./raw/GSE115978/suppl")

# 读表达矩阵：行是 gene，列是 cell
counts_path = data_dir / "GSE115978_counts.csv.gz"
matrix_df = pd.read_csv(counts_path, index_col=0)

# 读细胞注释
anno_path = data_dir / "GSE115978_cell.annotations.csv.gz"
anno_df = pd.read_csv(anno_path)

print("matrix:", matrix_df.shape)
print("anno:", anno_df.shape)
print(matrix_df.iloc[:5, :5])
print(anno_df.head())
```

```python
# 对齐表达矩阵和注释
anno_df["cells"] = anno_df["cells"].astype(str)
anno_df = anno_df.set_index("cells")

common_cells = matrix_df.columns.intersection(anno_df.index)

matrix_df = matrix_df.loc[:, common_cells]
anno_df = anno_df.loc[common_cells]

print(matrix_df.shape)
print(anno_df.shape)
print(matrix_df.columns.equals(anno_df.index))
```

```python
adata = sc.AnnData(X=matrix_df.T)
adata.obs = anno_df.copy()
adata.var_names = matrix_df.index.astype(str)

adata
```

```python
# 列名里的 . -> _
adata.obs.columns = adata.obs.columns.str.replace(".", "_", regex=False)

# 内容里的 . -> _
for col in adata.obs.select_dtypes(include="object").columns:
    adata.obs[col] = adata.obs[col].str.replace(".", "_", regex=False)

adata.obs.head()
```

```python
# include dataset column
adata.obs["dataset"] = 'GSE115978'
adata.obs["symptoms"] ='tumor'
adata.obs["gender"] = 'Unknown'
adata.obs["species"] = 'Human'
adata.obs["age"] = 'Unknown'
adata.obs["intervention"] = 'Unknown'
adata.obs["location"] = 'Unknown'
```

```python
adata.write_h5ad("./output_0531/GSE115978/GSE115978_raw.h5ad")
```
